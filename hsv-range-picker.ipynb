```python
from ipywidgets import interact, IntSlider
from matplotlib import pyplot as plt
import cv2
import numpy as np


IMAGE_PATH = 'dataset/trim/IMG_2426.jpeg'
ORIGINAL_IMG = cv2.imread(IMAGE_PATH)
ORIGINAL_IMG = cv2.cvtColor(ORIGINAL_IMG, cv2.COLOR_BGR2RGB)

HSV_IMG = cv2.cvtColor(ORIGINAL_IMG, cv2.COLOR_RGB2HSV)


def show2images(image1, image2, title1, title2):
    fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharex=True, sharey=True)
    ax = axes.ravel()

    ax[0].imshow(image1)
    ax[0].set_title(title1)

    ax[1].imshow(image2)
    ax[1].set_title(title2)

    for a in ax:
        a.set_axis_off()

    plt.tight_layout()
    plt.show()


def pickHSVMask(h_lo=0, h_hi=179, s_lo=0, s_hi=255, v_lo=0, v_hi=255):
    lower = np.array([h_lo, s_lo, v_lo], dtype=np.uint8)
    upper = np.array([h_hi, s_hi, v_hi], dtype=np.uint8)

    mask = cv2.inRange(HSV_IMG, lower, upper)
    colored = cv2.bitwise_and(ORIGINAL_IMG, ORIGINAL_IMG, mask=mask)
    show2images(ORIGINAL_IMG, colored, "Original", f"Masked (H[{h_lo},{h_hi}] S[{s_lo},{s_hi}] V[{v_lo},{v_hi}])")


# NOTE: best at
# low 8,130.10
# up 16,180,140
interact(pickHSVMask,
         h_lo=IntSlider(0, 0, 179),
         h_hi=IntSlider(179, 0, 179),
         s_lo=IntSlider(0, 0, 255),
         s_hi=IntSlider(255, 0, 255),
         v_lo=IntSlider(0, 0, 255),
         v_hi=IntSlider(255, 0, 255))
```
